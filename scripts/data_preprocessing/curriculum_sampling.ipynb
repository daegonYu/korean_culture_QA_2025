단답형 정답 작성 형식
- 하나의 질문에 대해 정답이 1개 존재하는 경우 : 허용되는 단어 : 한글, 영어, () 없음, 한자 없음
    - 예) 사과 (괄호나 한자 사용 금지)
- 하나의 질문에 대해 정답이 2개 이상 존재하는 경우 : ‘,’기준 나열
    - 예) 사과, 배
- 순서에 맞게 배열하세요 : ‘-’ 기준으로 나열
    - 예) ㄱ-ㄴ-ㄷ-ㄹ

```python
import pandas as pd
```

```python
training_df = pd.read_csv('/workspace/korean_culture_QA_2025/data/preprocessed/original_train_excluded_서술형.csv')
training_df
```

```python
answers_df = pd.read_csv('/workspace/korean_culture_QA_2025/data/preprocessed/curriculum_sampling_results_선다형_단답형_trillion-7B.csv')
answers_df
```

```python
df = pd.concat([training_df, answers_df], axis=1)
df
```

```python
import re
```

```python
def evaluate_short_answer(pred_answer, true_answer):

    """
    Calculate Exact Match score where true_data may contain multiple acceptable answers separated by #
    """
    correct = 0
    true_answer_list = true_answer.split('#')

    if any(pred_answer == ans for ans in true_answer_list):
        correct = 1
            
    return correct
```

```python
def evaluate_multiple_choice(pred_answer, true_answer):
    """선다형 평가"""
    # 숫자 추출
    # pred_nums = re.findall(r'\b[1-5]\b', pred_answer)
    pred_nums = re.findall(r'[1-5]', pred_answer)
    if pred_nums:
        pred = pred_nums[0]
    else:
        pred = pred_answer.strip()
    
    return 1 if pred == true_answer else 0
```

```python
n = 8
accs = []
for idx, row in df.iterrows():
    sum_score = 0
    if row['question_type'] == "선다형":
        for i in range(1, n+1):
            pred_answer = row[f'answer_{i}'].split('정답:')[-1].split('\n')[0].strip()
            pred_answer = pred_answer.replace('*', '')
            score = evaluate_multiple_choice(pred_answer, row['answer'])
            sum_score += score

    elif row['question_type'] == "단답형":
        for i in range(1, n+1):
            pred_answer = row[f'answer_{i}'].split('정답:')[-1].split('\n')[0].strip()
            pred_answer = pred_answer.replace('*', '').replace(' ','')
            score = evaluate_short_answer(pred_answer, row['answer'].replace(' ',''))
            sum_score += score
    acc = sum_score/n
    accs.append(acc)
df['accuracy'] = accs
```
